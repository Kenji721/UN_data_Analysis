# Exploratory Data Analysis for GDP Classification

## 1. Target Variable Analysis
Let's analyze the distribution and characteristics of our target variable `gdp_class`.

```python
# Target variable analysis
plt.figure(figsize=(15, 5))

# GDP class distribution
plt.subplot(1, 3, 1)
gdp_counts = df['gdp_class'].value_counts().sort_index()
plt.bar(gdp_counts.index, gdp_counts.values, alpha=0.7, color='skyblue')
plt.title('GDP Class Distribution')
plt.xlabel('GDP Class')
plt.ylabel('Count')
for i, v in enumerate(gdp_counts.values):
    plt.text(i, v + 2, str(v), ha='center')

# GDP per capita distribution by class
plt.subplot(1, 3, 2)
df.boxplot(column='GDP per capita (current US$)', by='gdp_class', ax=plt.gca())
plt.title('GDP per Capita by Class')
plt.suptitle('')  # Remove automatic title

# GDP per capita histogram
plt.subplot(1, 3, 3)
plt.hist(df['GDP per capita (current US$)'].dropna(), bins=50, alpha=0.7, color='lightcoral')
plt.title('GDP per Capita Distribution')
plt.xlabel('GDP per Capita (US$)')
plt.ylabel('Frequency')
plt.yscale('log')  # Log scale due to wide range

plt.tight_layout()
plt.show()

# Show GDP ranges for each class
print("GDP per capita ranges by class:")
for cls in sorted(df['gdp_class'].dropna().unique()):
    gdp_values = df[df['gdp_class'] == cls]['GDP per capita (current US$)'].dropna()
    print(f"Class {int(cls)}: ${gdp_values.min():,.0f} - ${gdp_values.max():,.0f} (median: ${gdp_values.median():,.0f})")
```

## 2. Missing Data Analysis
Understanding missing data patterns is crucial for classification.

```python
# Missing data analysis
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)

# Missing Data Pattern and Percentage - Side by Side
plt.figure(figsize=(18, 6))

# Missing data heatmap
plt.subplot(1, 2, 1)
# Select columns with most missing data for visualization
top_missing_cols = missing_data.head(20).index
sns.heatmap(df[top_missing_cols].isnull(), cbar=True, yticklabels=False, cmap='viridis')
plt.title('Missing Data Pattern (Top 20 columns)')
plt.xlabel('Features')

# Missing data bar plot
plt.subplot(1, 2, 2)
missing_pct = (missing_data.head(15) / len(df) * 100)
missing_pct.plot(kind='bar')
plt.title('Missing Data Percentage (Top 15)')
plt.ylabel('Missing %')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

print("Features with most missing data:")
print(missing_data.head(10))
```

```python
# Missing Data by GDP Class and Feature Completeness Analysis
plt.figure(figsize=(15, 6))

# Missing data by GDP class
plt.subplot(1, 2, 1)
missing_by_class = df.groupby('gdp_class').apply(lambda x: x.isnull().sum().sum())
missing_by_class.plot(kind='bar', color='coral')
plt.title('Total Missing Values by GDP Class')
plt.xlabel('GDP Class')
plt.ylabel('Missing Count')
plt.xticks(rotation=0)

# Feature completeness
plt.subplot(1, 2, 2)
completeness = (df.count() / len(df) * 100).sort_values()
plt.hist(completeness, bins=20, alpha=0.7, color='lightgreen')
plt.title('Feature Completeness Distribution')
plt.xlabel('Completeness %')
plt.ylabel('Number of Features')

plt.tight_layout()
plt.show()

print(f"Features with >50% missing data: {len(completeness[completeness < 50])}")
print(f"Features with >90% completeness: {len(completeness[completeness > 90])}")
```

## 3. Feature Analysis by GDP Class
Let's analyze how different features vary across GDP classes.

```python
# Select key features for analysis - focusing on complete and potentially discriminative features
key_features = [
    'Population (000, 2024)', 'Pop. density (per km2, 2024)', 'Surface area (km2)',
    'Sex ratio (m per 100 f)', 'Life expectancy at birth (years)', 
    'Education: Government expenditure (% of GDP)', 'Health: Government expenditure (% of GDP)',
    'Individuals using the Internet (%)', 'Military expenditure (% of GDP)'
]

# Filter features that exist in the dataset
existing_key_features = [f for f in key_features if f in df.columns]
print(f"Analyzing {len(existing_key_features)} key features:")
print(existing_key_features)

# Statistical analysis by GDP class
plt.figure(figsize=(20, 15))

for i, feature in enumerate(existing_key_features[:6]):  # Limit to 6 for readability
    plt.subplot(2, 3, i+1)
    
    # Box plot by GDP class
    df_clean = df[df[feature].notna() & df['gdp_class'].notna()]
    if len(df_clean) > 0:
        sns.boxplot(data=df_clean, x='gdp_class', y=feature)
        plt.title(f'{feature} by GDP Class')
        plt.xticks(rotation=45)
    else:
        plt.text(0.5, 0.5, 'No complete data', ha='center', va='center')
        plt.title(f'{feature} (No Data)')

plt.tight_layout()
plt.show()
```

```python
# Correlation analysis
# Select numeric features with less than 70% missing data for correlation analysis
numeric_cols = df.select_dtypes(include=[np.number]).columns
complete_features = []
for col in numeric_cols:
    if col != 'gdp_class' and df[col].notna().sum() / len(df) > 0.3:  # At least 30% complete
        complete_features.append(col)

print(f"Found {len(complete_features)} features with >30% completeness for correlation analysis")

# Calculate correlation with GDP class
correlations = []
for feature in complete_features:
    clean_data = df[[feature, 'gdp_class']].dropna()
    if len(clean_data) > 10:  
        corr = clean_data[feature].corr(clean_data['gdp_class'])
        correlations.append((feature, corr, len(clean_data)))

# Sort by absolute correlation
correlations = sorted(correlations, key=lambda x: abs(x[1]), reverse=True)

# Display top correlations
print("\nTop 15 features most correlated with GDP class:")
for i, (feature, corr, n_obs) in enumerate(correlations[:15]):
    print(f"{i+1:2d}. {feature[:50]:<50} | Corr: {corr:6.3f} | N: {n_obs:3d}")

# Create correlation heatmap for top features
if len(correlations) >= 10:
    top_features = [item[0] for item in correlations[:10]] + ['gdp_class']
    corr_matrix = df[top_features].corr()
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', center=0, fmt='.2f',
                square=True, cbar_kws={'label': 'Correlation'})
    plt.title('Correlation Matrix: Top 10 Features vs GDP Class')
    plt.tight_layout()
    plt.show()
```

## 4. Feature Engineering and Data Preparation
Let's prepare the data for classification by handling missing values and creating derived features.

```python
# Feature selection for classification
# Select features with high correlation and good completeness
selected_features = [
    'Employment in agriculture (% of employed)',
    'Employment in services (% employed)', 
    'Economy: Agriculture (% of Gross Value Added)',
    'Fertility rate, total (live births per woman)',
    'Education: Upper Sec. gross enrol. ratio - Male (per 100 pop.)',
    'Education: Upper Sec. gross enrol. ratio - Female (per 100 pop.)',
    'Health: Physicians (per 1 000 pop.)',
    'Under five mortality rate (per 1000 live births)',
    'Urban population (% of total population)',
    'Economy: Services and other activity (% of GVA)',
    'Population growth rate (average annual %)',
    'Employment in industry (% of employed)'
]

# Create dataset for classification
df_class = df[selected_features + ['gdp_class']].copy()

# Remove rows with missing target variable
df_class = df_class.dropna(subset=['gdp_class'])

print(f"Dataset shape after removing missing targets: {df_class.shape}")
print(f"Missing values per feature:")
print(df_class.isnull().sum())

# Calculate completeness for each feature
completeness = (df_class.count() / len(df_class) * 100).round(1)
print(f"\nFeature completeness (%):")
for feature, comp in completeness.items():
    if feature != 'gdp_class':
        print(f"{feature}: {comp}%")
```

```python
# Create different datasets based on completeness requirements
# Option 1: Use only complete cases (might reduce sample size significantly)
df_complete = df_class.dropna()
print(f"Complete cases dataset: {df_complete.shape}")

# Option 2: Use features with >80% completeness
high_complete_features = [feature for feature in selected_features 
                         if (df_class[feature].count() / len(df_class)) > 0.8]
high_complete_features.append('gdp_class')
df_high_complete = df_class[high_complete_features].copy()
print(f"High completeness features ({len(high_complete_features)-1}): {high_complete_features[:-1]}")
print(f"High completeness dataset: {df_high_complete.shape}")

# Option 3: Most complete features for maximum sample size
most_complete_features = [
    'Fertility rate, total (live births per woman)',
    'Under five mortality rate (per 1000 live births)', 
    'Economy: Services and other activity (% of GVA)',
    'Population growth rate (average annual %)',
    'Economy: Agriculture (% of Gross Value Added)',
    'gdp_class'
]
df_most_complete = df_class[most_complete_features].dropna()
print(f"Most complete dataset: {df_most_complete.shape}")

# Visualize sample sizes vs feature completeness tradeoff
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Sample sizes comparison
datasets = ['Complete Cases', 'High Completeness\n(>80%)', 'Most Complete\n(>98%)']
sample_sizes = [df_complete.shape[0], df_high_complete.dropna().shape[0], df_most_complete.shape[0]]
feature_counts = [len(selected_features), len(high_complete_features)-1, len(most_complete_features)-1]

ax1.bar(datasets, sample_sizes, alpha=0.7, color=['coral', 'skyblue', 'lightgreen'])
ax1.set_title('Sample Sizes by Completeness Strategy')
ax1.set_ylabel('Number of Observations')
for i, v in enumerate(sample_sizes):
    ax1.text(i, v + 5, str(v), ha='center')

ax2.bar(datasets, feature_counts, alpha=0.7, color=['coral', 'skyblue', 'lightgreen'])
ax2.set_title('Number of Features by Strategy')
ax2.set_ylabel('Number of Features')
for i, v in enumerate(feature_counts):
    ax2.text(i, v + 0.1, str(v), ha='center')

plt.tight_layout()
plt.show()

print(f"\nClassification strategy recommendations:")
print(f"1. Complete cases: {df_complete.shape[0]} samples, {len(selected_features)} features - Good for robust analysis")
print(f"2. High completeness: {df_high_complete.dropna().shape[0]} samples, {len(high_complete_features)-1} features - Balanced approach")
print(f"3. Most complete: {df_most_complete.shape[0]} samples, {len(most_complete_features)-1} features - Maximum sample size")
```

## 5. EDA Summary and Classification Insights

### Key Findings:

1. **Target Variable**: GDP class is well-balanced with ~158 observations per class (0-3)
2. **Most Predictive Features**:
   - Employment in agriculture (% of employed) - strong negative correlation (-0.801)
   - Employment in services (% employed) - strong positive correlation (0.786) 
   - Economy: Agriculture (% of Gross Value Added) - negative correlation (-0.745)
   - Fertility rate - negative correlation (-0.733)
   - Education indicators - positive correlations (~0.73)

3. **Data Quality**:
   - 18.22% overall missing data
   - 8 features with >50% missing data
   - 37 features with >90% completeness

4. **Classification Strategy Options**:
   - **Robust approach**: 297 complete cases with 12 features
   - **Balanced approach**: 457 samples with 10 high-completeness features  
   - **Maximum data**: 621 samples with 5 most complete features

### Recommendations for Classification:
1. Start with the balanced approach (457 samples, 10 features) for optimal trade-off
2. Key features show clear economic development patterns
3. Consider feature engineering (ratios, interactions) for better performance
4. Handle missing data with imputation if needed for more complex models
