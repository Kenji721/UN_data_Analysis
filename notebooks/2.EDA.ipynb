```python
import pandas as pd
```

```python
df_general_info = pd.read_csv('../data/processed/general_info_clean.csv')
df_general_info.head()
```

```python
df_general_info.info()
```

```python
df_general_info.isnull().sum()
```

```python
# List countries with missing data in specific columns

missing_exchange = df_general_info[df_general_info['Exchange rate (per US$)'].isna()]['Country']
print(f"Countries missing exchange rate data ({len(missing_exchange)} total):")
print(missing_exchange.tolist())

print("\n" + "="*50)

missing_nationalcur = df_general_info[df_general_info['National currency'].isna()]['Country']
print(f"Countries missing National currency data ({len(missing_nationalcur)} total):")
print(missing_nationalcur.tolist())

print("\n" + "="*50)

missing_popdense = df_general_info[df_general_info['Pop. density (per km2, 2024)'].isna()]['Country']
print(f"Countries missing Pop. density (per km2, 2024) data ({len(missing_popdense)} total):")
print(missing_popdense.tolist())

print("\n" + "="*50)

missing_pops = df_general_info[df_general_info['Population (000, 2024)'].isna()]['Country']
print(f"Countries missing Population (000, 2024) data ({len(missing_pops)} total):")
print(missing_pops.tolist())

print("\n" + "="*50)

missing_sexratio = df_general_info[df_general_info['Sex ratio (m per 100 f)'].isna()]['Country']
print(f"Countries missing Sex ratio (m per 100 f) data ({len(missing_sexratio)} total):")
print(missing_sexratio.tolist())

print("\n" + "="*50)

missing_surfarea = df_general_info[df_general_info['Surface area (km2)'].isna()]['Country']
print(f"Countries missing Surface area (km2) data ({len(missing_surfarea)} total):")
print(missing_surfarea.tolist())

print("\n" + "="*50)

missing_date = df_general_info[df_general_info['UN membership date'].isna()]['Country']
print(f"Countries missing UN membership date data ({len(missing_date)} total):")
print(missing_date.tolist())

print("\n" + "="*50)

missing_capcitypop = df_general_info[df_general_info['Capital_city_pop'].isna()]['Country']
print(f"Countries missing Capital_city_pop data ({len(missing_capcitypop)} total):")
print(missing_capcitypop.tolist())

print("\n" + "="*50)

```

1. Exchange Rate
    - State of Palestine has no piblic national currency  
2. National Currency
    - **Palestine** has no public national currency 
3. pop density, population, sex ration
    - **Channel Islands** and **Holy See** has no information regarding their population
4. Surface area
    - Counties missing these data are Depandant territories/ Special Jurisdiction
5. Missing capital city pop
    - Tokelau has no capital city
