```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/general_info_clean.csv')
df.head()
```

```python
# Fill missing 2024 data with older data
df['Capital_city_pop_merged'] = df['Capital city pop. (000, 2024)'].fillna(df['Capital city pop. (000)'])
df = df.drop(columns=['Capital city pop. (000)', 'Capital city pop. (000, 2024)'])
```

```python
df.isnull().sum()
```

```python
# List countries with missing data in specific columns

# Example 1: Countries missing exchange rate data
if 'Exchange rate (per US$)' in df.columns:
    missing_exchange = df[df['Exchange rate (per US$)'].isna()]['Country']
    print(f"Countries missing exchange rate data ({len(missing_exchange)} total):")
    print(missing_exchange.tolist())

print("\n" + "="*50)

# Example 2: Countries missing surface area data
if 'Surface area (km2)' in df.columns:
    missing_surface = df[df['Surface area (km2)'].isna()]['Country']
    print(f"Countries missing surface area data ({len(missing_surface)} total):")
    print(missing_surface.tolist())

print("\n" + "="*50)

# Example 3: Show full rows for countries with missing population density
if 'Pop. density (per km2, 2024)' in df.columns:
    missing_density = df[df['Pop. density (per km2, 2024)'].isna()]
    print(f"Countries missing population density data ({len(missing_density)} total):")
    print(missing_density[['Country', 'Pop. density (per km2, 2024)', 'Population (000, 2024)']].head(10))

print("\n" + "="*50)

# Example 4: Check missing data for merged capital city population
missing_capital_pop = df[df['Capital_city_pop_merged'].isna()]['Country']
print(f"Countries missing capital city population data ({len(missing_capital_pop)} total):")
print(missing_capital_pop.tolist())
```
