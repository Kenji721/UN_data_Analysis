```python
# Statistical tests and analysis
print("=== EXCHANGE RATE STATISTICAL ANALYSIS ===\n")

# Basic statistics
rates_clean = df['Exchange rate (per US$)'].dropna()
print(f"Total countries with exchange rate data: {len(rates_clean)}")
print(f"Mean exchange rate: {rates_clean.mean():.2f}")
print(f"Median exchange rate: {rates_clean.median():.2f}")
print(f"Standard deviation: {rates_clean.std():.2f}")
print(f"Coefficient of variation: {(rates_clean.std()/rates_clean.mean())*100:.2f}%")

print(f"\nPercentiles:")
for p in [25, 50, 75, 90, 95, 99]:
    print(f"{p}th percentile: {rates_clean.quantile(p/100):.2f}")

# Skewness analysis
from scipy import stats
skewness = stats.skew(rates_clean)
print(f"\nSkewness: {skewness:.2f}")
if skewness > 1:
    print("Distribution is highly right-skewed (long tail towards high values)")
elif skewness > 0.5:
    print("Distribution is moderately right-skewed")
else:
    print("Distribution is relatively symmetric")
```
